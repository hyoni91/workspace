### 결측치제거

결측치 : 값이 없는 것    (Not a Number) NaN

```python
import pandas

df = pandas.read_excel('test2.xlsx')

df
```

```python
# NaN 데이터를 다른 문자로 변경
# fillna (na를 fill하겠다.)
# 오류 방지를 위해서 하기 코드로는 저장 되지 않는다. 
df.fillna('없음') 

# 그렇다면 확정은? inplace=True
df.fillna('없음',inplace=True)

df
```

```python
# 여러 컬럼에 NaN가 있으면 특정 컬럼을 지정해줘야 한다.
# 연습을 위해 다시 엑셀 로드

df = pandas.read_excel('test2.xlsx')
df

# 학급의 모든 데이터를 nan으로 변경
import numpy
df['grand'] = numpy.nan
df

# 학급 NaN = 미지정
# 취미 NaN = 없음
df['grand'].fillna('미지정')
df
# 확정 지어줘야함!!
df['grand'].fillna('미지정',inplace=True)
df

# 새롭게 바뀐 방법? nan 컬럼명을 지정하고 바꾸기?
df['hobby'] = df['hobby'].fillna('없음')
df
```

```python
# 통상적으로 NaN이 있는 경우 오류 데아터일 경우가 많기 때문에
# 결측치를 제거한다(해당 행 또는 열 모두 삭제해야 한다.)


# 연습을 위해 리로드
df = pandas.read_excel('test2.xlsx')
df


```

```python
# 결측치 제거 방법 (NaN가 포함된 행 또는 열 삭제)
# 아래코드는 열 삭제
df.dropna()
# 확정을 지으려면 inplace=True
df
```

```python
# dropna의 매개변수 정보
# 자주 쓰는 매개변수 axis, how 
# axis : ex)  df.dropna(axis=??, inplace=True)
# axis = 'index' : NaN 데이터를 기준으로 행을 삭제한다. (default)
# axis = 'colums' : NaN 데이터를 기준으로 열을 삭제한다.
df.dropna(axis='columns')

# how : ex) df.dropna(how=??, inplace=True)
# how = 'any' : NaN값이 하나라도 있으면 삭제 (default) 
# how = 'all' : 모든 값이 NaN이면 삭제

df.dropna(how='any')  # axis값을 따로 지정하지 않았기 때문에 axis의 디폴트 값인 행을 삭제하게 된다.

# 결측치 값이 모두 NaN이면 해당 열 삭제
df.dropna(axis='columns', how='any')

```
