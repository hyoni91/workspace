### 엑셀파일로 저장

```python
data = {
  'name' : ['김김','이씨','박둘','최자','정지우'],
  'grand' : ['하늘반','바다반','숲반','우미반','소라반'],
  'korSco' : [90,80,70,100,90],
  'engSco' : [70,80,80,70,90],
  'avg' : [80.0,80.0,75.0,85.0,90.0],
  'hobby' : ['독서','','','요가','축구'] 
}

data
```

```python
import pandas

df = pandas.DataFrame(data)

df
```

```python
# 데이터프레임을 엑셀파일로 저장
# 저장되는 위치는 현재 파이썬을 연 파일안에 만들어 진다.
df.to_excel('test1.xlsx')
```

```python
# 엑셀 파일을 불러오기

df_excel = pandas.read_excel('test1.xlsx')

df_excel

# 인덱스도 엑셀에 등록이 되어 있기때문에 unnamed로 컬럼으로 들고오게됨
# 판다스는 엑셀파일을 들고오면서 다시 인덱스를 만들어서 들어옴 
# 빈 문자열은 NaN오류로??
```
