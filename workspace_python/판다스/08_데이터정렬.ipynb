### 데이터 정렬

```python
import pandas

df = pandas.read_excel('test2.xlsx')

df
```

```python
# 컬럼 기준 데이터 정렬 또는 인덱스 기준으로 정렬 가능

# 인덱스추가
index_list = ['1번','2번','3번','4번','5번']
df.index = index_list
df.index.name = 'stuNum'
df
```

```python
# 열 데이터로 정렬
df
# 평균 기준 오름차순(defult)
df.sort_values('avg')
df.sort_values('avg', ascending=True)

# 국어점수 기준 내림차순
df.sort_values('korSco', ascending=False)

```

```python
# 평균 기준 내림차순 정렬후, 평균이 같으면 국어 점수 기준 내림차순 정렬

df.sort_values(['avg','korSco'], ascending=[False,False])
```

```python
# 인덱스 기준 정렬

df.sort_index(ascending=True)
df.sort_index(ascending=False)

```
